### 그리드 설정 (grid)


```python
def draw_line(grid=False, style='b-'):
  plt.title("나이와 재산의 상관관계")
  plt.plot(
    range(10,50,10),
    [1,4,9,16],
    style
    )
  plt.xlabel('나이')
  plt.ylabel('재산')
  plt.grid(grid)
  plt.show()
```

```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
draw_line(grid=True)
draw_line(grid=False)
```

### 여러 개의 선 그어보기

```python
t = np.arange(0., 5., 0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title("라인 플롯에서 여러개의 선")
plt.plot(
    t, t, 'r--',
    t, t**2, 'bs:',
    t, t**3, 'g^-',
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title("여러개의 plot 명령으로 한 그림에 겹쳐서 표현")
plt.plot((np.arange(4)+1)**2)
plt.plot((np.arange(4)-1)**0.3)
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X = np.linspace(-np.pi, np.pi, 256)
```

```python
# 코사인, 사인 곡선
C, S = np.cos(X), np.sin(X)
```

```python
plt.title('삼각함수')
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
plt.title('삼각함수')
plt.plot(X,C, label = 'cosine')
plt.plot(X,S, label = 'sine')
plt.legend(loc=2)
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2,1,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")
plt.subplot(2,1,2)
plt.plot(np.arange(10),np.arange(10)**3, 'r--')
plt.title("차트2")

plt.tight_layout() #위아래가 겹치지않게 해줌
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
plt.subplot(1,2,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")
plt.subplot(1,2,2)
plt.plot(np.arange(10),np.arange(10)**3, 'r--')
plt.title("차트2")

plt.tight_layout() #위아래가 겹치지않게 해줌
plt.show()
```

```python
# 2x2 플롯 배치
plt.figure(figsize = (15,12))
plt.subplot(2,2,1)
plt.plot(np.arange(5)**2)
plt.title("차트1")
plt.xlabel("x축")
plt.ylabel("y축")
plt.subplot(2,2,2)
plt.plot(np.arange(10),np.arange(10)**3, 'r--')
plt.title("차트2")
plt.subplot(2,2,3)
plt.plot(np.arange(10),np.arange(10)**3, 'b--')
plt.title("차트3")
plt.subplot(2,2,4)
plt.plot(np.arange(10),np.arange(10)**3, 'y--')
plt.title("차트4")

plt.tight_layout() #위아래가 겹치지않게 해줌
plt.show()
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2,3,1]
x = np.arange(3)
xlabel=['가','나','다']
plt.title("바 차트")
plt.xticks(x, xlabel) #x축에 이름 붙이기
plt.yticks(sorted(y)) #y축에 기준이 되는이름 붙여주기
plt.bar(x,y)
plt.show()
```

```python
y = [2,3,1]
x = np.arange(3)
xlabel=['가','나','다']
plt.title("바 차트")
plt.yticks(x, xlabel) #x축에 이름 붙이기
plt.xticks(sorted(y)) #y축에 기준이 되는이름 붙여주기
plt.barh(x,y) ####가로로 바 차트 그리기!! horizontal/// plt.barh()
plt.show()
```

```python
people = ['james', 'john', 'kate', 'alice']
y_pos = np.arange(4)
performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar-h Chart')
plt.grid(True)
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x축')
plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ['호박파이', '사과파이','호두파이','엄마손파이']
sizes = [15,30,45,10]
colors = ['yellow','gold','lightblue','lightcoral']
explode = (0,0.1,0,0)
plt.title("파이차트")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
# kind = 종류 => 해당 종류 plot 교체
iris['sepal_length'][:20].plot(kind='bar', rot=45, ylim=(3.5,6))
plt.title('꽃받침의 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.show()
```
